### shuffling of data

```python
data_shuffled=df.sample(frac=1).reset_index(drop=True)
```

```python
data_shuffled
```

```python
labels=data_shuffled.iloc[:,-1]
data_shuffled_=data_shuffled.drop(data_shuffled.columns[[25]],axis=1)
```

#### Preprocessing of shuffled data

```python
# manually scaling of data row-wise from 0 to 
from sklearn.preprocessing import minmax_scale
for i in range (3000):
    min_row=min(data_shuffled.iloc[i,:])
    max_row=max(data_shuffled.iloc[i,:])
    for j in range (25): 
        data_shuffled.iloc[i,j]=(data_shuffled.iloc[i,j]-min_row)/(max_row-min_row)
```

#### Binary Classification between feature value 0 and 1 for first 10 features only

```python
## Selecting only rows that has feature value 0 and 1 
data_binary=data_shuffled[np.logical_or(data_shuffled[25]==0 , data_shuffled[25]==1)]
```

```python
## selecting only 10 features
label_binary=data_binary.iloc[:,-1]
feature_binary=data_binary.iloc[:,:10]
```

```python
feature_binary
```

```python
## splitting data into train and test. Since our data is small we will do cross validation on train data
x_train_bi,x_test_bi=feature_binary[:500],feature_binary[500:]
y_train_bi,y_test_bi=label_binary[:500],label_binary[500:]
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_validate
#clf=SVC()
#clf.fit(x_train_bi, y_train_bi)
#cv_results = cross_validate(clf, x_train_bi, y_train_bi, cv=5)
#print(cv_results['test_score'])
```

```python
model_params={'SVC':{
                    'model':SVC(),
                    'params':{'C':[0.01,0.1,1,10],'kernel':['linear', 'poly', 'rbf', 'sigmoid', 'precomputed'],'degree':[1,2,3]}
                      },
            
             }
```

```python
from sklearn.model_selection import GridSearchCV
scores=[]
for model_name,mp in model_params.items():
    clf=GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
    clf.fit(x_train_bi, y_train_bi)
    scores.append({
        'model':model_name,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    })
```

```python
y_pred=clf.predict(x_test_bi)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, r2_score
print(confusion_matrix(y_test_bi,y_pred))
print(r2_score(y_test_bi,y_pred))
print(classification_report(y_test_bi,y_pred))
```
